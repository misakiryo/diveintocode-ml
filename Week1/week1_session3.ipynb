【問題】栗まんじゅうが太陽系を覆う日


1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。


栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。


ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。

```python
#授業後に1人で解いたもの



import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#栗まんじゅう体積 10-4[m3]　https://www.sekkachi.com/entry/bibin_consideration
#太陽系体積  3*10**80[m3] 引用上記URL


def cover_time(increase_vol, source_vol):
  minutes = 0
  list_increase_vol = [increase_vol]
  
  while list_increase_vol[-1] <= source_vol:
    minutes += 5
    list_increase_vol.append(list_increase_vol[-1]*2)
  return minutes, list_increase_vol


kuri_manju_vol = 10**-4
solar_system_vol = 3*(10**80)
increase_vol = kuri_manju_vol
source_vol = solar_system_vol
minutes, list_increase_vol = cover_time(increase_vol, source_vol)
print('{}回倍で{}分後に覆われる'.format(int(minutes/5), minutes))
```

```python
#グラフで表示する
#解答参照しながら（x軸の値の出し方)


plt.figure(figsize=(13, 8))
plt.title("cover_date", {'fontsize':20})
plt.xlabel("minutes", {'fontsize':20})
plt.ylabel("increase_volume", {'fontsize':20})
plt.hlines(source_vol, 0,minutes, "red", linestyles='dashed', label='solar system')
plt.grid(True)
plt.plot(range(0,minutes+5,5), list_increase_vol, label = 'kuri_manju', marker = 'o')
plt.legend()
plt.show()
```

```python
#解答写しながら分析・・・
#whileの使い方、定義づけの種類、分数の扱い方
#グラフ→ x軸とy軸の表記方法、一定の値の線の引き方

import math
import matplotlib.pyplot as plt
%matplotlib inline

r_solar_system = 120*149597870700
volume_of_solar_system = (4/3)*math.pi*r_solar_system**3
print("太陽系の体積は{}m^3".format(volume_of_solar_system))

volume_of_kurimanju = (60*0.001)*(40*0.001)*(30*0.001)
print("栗まんじゅうの体積は{}m^3".format(volume_of_kurimanju))

def compute_baybayin_problem(volume_of_space, volume_of_object_baybayined, minutes_per_count=5):
  list_volumes  = [volume_of_object_baybayined]
  count = 0
  while True:
      count += 1
      list_volumes.append(list_volumes[-1]*2)
      if list_volumes[-1]>=volume_of_space:
          break
  minutes = count*minutes_per_count
  return minutes, count, list_volumes

minutes_per_count = 5

r_solar_system = 120*149597870700
volume_of_solar_system = (4/3)*math.pi*r_solar_system**3

volume_of_kurimanju = (60*0.001)*(40*0.001)*(30*0.001)

volume_of_space = volume_of_solar_system
volume_of_object_baybayined = volume_of_kurimanju

minutes, count, list_volumes = compute_baybayin_problem(volume_of_space, volume_of_object_baybayined, minutes_per_count=minutes_per_count)

print("{}回倍になる{}分で覆われる".format(count, minutes))

```

```python
plt.title("baybayin")
plt.xlabel("minutes")
plt.ylabel("volumes[m^3]")
plt.plot(range(0, minutes + minutes_per_count, minutes_per_count), list_volumes, label="kurimanju", marker="o")
# 太陽系の体積を赤い点線で表示
plt.hlines(volume_of_space, 0,minutes, "red", linestyles='dashed', label='solar system')
plt.legend(loc="best") # 凡例を最適な位置に表示
plt.show()
```
