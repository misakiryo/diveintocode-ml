【問題1】乱数の作成

平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。


[1.0　0.8]

[0.8　1.0]


```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt

mean = [-3, 0]
cov = [[1.0,0.8] , [0.8,1.0]]

#data_1が生成される前に置くと固定されるみたい
np.random.seed(0)

data_1 = np.random.multivariate_normal(mean, cov, size= 500)
data_1.shape
```

【問題2】散布図による可視化

問題1で作成したデータ点を散布図により可視化してください。


散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
data_1_a,data_1_b = np.random.multivariate_normal(mean, cov, size=500).T

plt.title('scatter_plot')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(data_1_a, data_1_b, c="r", marker="*", alpha=0.5, linewidths="0.5")
plt.grid(True)
```

【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。



ヒストグラムはplt.hist()を使うことで描けます。



各次元ごとに作成し、2つのヒストグラムとしてください。


それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。



```python
#1種類目

plt.title('data_1_a')
plt.xlabel('x1')
plt.ylabel('y1')
plt.xlim(-5.5,2.5)
plt.ylim(0,30)
plt.hist(data_1_a, 65)
```

```python
#2種類目

plt.title('data_1_b')
plt.xlabel('x2')
plt.ylabel('y2')
plt.xlim(-5.5,2.5)
plt.ylim(0,30)
plt.hist(data_1_b, 65)
```

【問題4】データの追加

新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。


[1.0 0.8]

[0.8 1.0]

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題4のものは1を表示してください。



```python
np.random.seed(0)

mean_2 = [0, -3]
cov_2 = [[1.0,0.8] , [0.8,1.0]]
data_2 = np.random.multivariate_normal(mean_2, cov_2, size=500)
data_2_a, data_2_b = np.random.multivariate_normal(mean_2, cov_2, size=500).T
```

```python
#散布図の可視化、凡例の表示

fig=plt.figure(figsize=(15,7))
plt.title('scatter')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(data_1_a,data_1_b, c="r", marker="*", alpha=0.5, linewidths="0.1", label ='0') #問題1
plt.scatter(data_2_a,data_2_b, c="y", marker=".", alpha=0.5, linewidths="0.3", label = '1') #問題4
plt.legend(loc = 'best')
plt.grid(True)
```

【問題6】ラベル付け

ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。


機械学習に使用するデータセットはこのような形になっていることが多いです。


《作成例》


データが3個+3個の場合は以下のような(6, 3)のndarrayになります。


Image from Gyazo

```python
#結合したデータの前のデータ毎に、ラベルを列に追加してから、再結合
#結合データ状態からのラベルを列に追加のやり方は分からない

data_1_zero = np.insert(data_1, 2, 0, axis=1)
data_2_one = np.insert(data_2, 2, 1, axis=1)
join_data_new_col = np.vstack((data_1_zero, data_2_one))
join_data_new_col.shape
```
